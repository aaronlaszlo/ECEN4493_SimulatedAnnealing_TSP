```python
from anneal import SimAnneal
import matplotlib.pyplot as plt
import random




def read_coords(path):
    coords = []
    with open(path, "r") as f:
        for line in f.readlines():
            line = [float(x.replace("\n", "")) for x in line.split(" ")]
            coords.append(line)
    return coords


def generate_random_coords(num_nodes):
    return [[random.uniform(-1, 1), random.uniform(-1, 1)] for i in range(num_nodes)]


if __name__ == "__main__":
    coords = read_coords("coord.txt")  # generate_random_coords(100)
    initsa = SimAnneal(coords, stopping_iter = 0)
    initsa.anneal()
    initsa.visualize_routes()
    sa = SimAnneal(coords, stopping_iter=10000)
    sa.anneal()
    sa.visualize_routes()
    sa.plot_learning()
```
